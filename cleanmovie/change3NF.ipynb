```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv('movie_clean.csv')
data.head(10)
```

```python
data = data[['movieId', 'title', 'adult', 'genres', 'homepage', 'original_language', 'overview', 'release_date', 'runtime', 'id',  'imdbId']]
```

```python
data.head(10)
```

```python
data.shape
```

```python
data['homepage'] = data['homepage'].replace(np.nan, 'missing')
data['adult'] = data['adult'].replace(np.nan, 'False')
data['original_language'] = data['original_language'].replace(np.nan, 'en')
data['overview'] = data['overview'].replace(np.nan, 'not available')
data['release_date'] = data['release_date'].replace(np.nan, '1926-08-17')
data['runtime'] = data['runtime'].replace(np.nan, '90')
```

```python
data.head(10)
```

```python
data.to_csv('movie_modified.csv', index=False)
```

```python
data.dtypes
```

```python
import ast
file = open('movie_modified.csv',encoding="utf8")
data = csv.reader(file)
i = 0
row_list = []
h = {}
for row in data:
    if i > 0:
        dictionary = ast.literal_eval(row[3])
        for v in dictionary:
            if len(v) > 1:
                h[v['id']] = v['name']
    i += 1
```

```python
print(h)
```

```python
row_list.append(['id', 'name'])
for key, value in h.items():
    row_list.append([key, value])
```

```python
with open('genres.csv', 'w',encoding="utf8", newline='') as file:
    writer = csv.writer(file)
    writer.writerows(row_list)
```

```python
file = open('movie_modified.csv',encoding="utf8")
data = csv.reader(file)
i = 0
row_list = []
for row in data:
    if i > 0:
        dictionary = ast.literal_eval(row[3])
        for v in dictionary:
            row_list.append([row[0], v['id']])
    else:    
        row_list.append([row[0], row[3]])
    i += 1
with open('moviegenres.csv', 'w',encoding="utf8", newline='') as file:
    writer = csv.writer(file)
    writer.writerows(row_list)
```
