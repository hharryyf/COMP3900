```python
import pandas as pd
import numpy as np
import csv
```

```python
def encodeid(ss):
    st = str(ss)
    while (len(st) < 7):
        st = '0' + st
    st = 'tt' + st
    return st
```

```python
encodeid(0)
```

```python
file = open('links_small.csv')
data = csv.reader(file)
i = 0
row_list = []
for row in data:
    row.pop()
    if i > 0:
        row[1] = encodeid(row[1])
    row_list.append(row)
    i += 1
```

```python
with open('clear_link.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(row_list)
```

```python
data = pd.read_csv('clear_link.csv')
data.head(10)
```

```python
data2 = pd.read_csv('movies_metadata.csv')
data.dropna(subset=['imdbId'], how='all', inplace=True)
ret = pd.concat([data, data2], axis=1)
ret.dropna(subset=['movieId'], how='all', inplace=True)
ret.head(10)
```

```python
ret = ret.astype({'movieId': 'int32'})
```
